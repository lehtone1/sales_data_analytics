### Filtering double campaigns

```python
rule_tv = df['tv_grps'] != 0
rule_radio = df['radio_grps'] != 0
rule_digital = df['digital_grps'] != 0
```

```python
tv_index = df[rule_tv].index.values
radio_index = df[rule_radio].index.values
digital_index = df[rule_digital].index.values
```

```python
def isin_extended(listx1, listx2):
    in1 = np.isin(listx1, listx2[0], invert=True)
    in2 = np.isin(listx1, listx2[1], invert=True)
    size = listx1.size
    in_combined = np.ones(size, dtype=bool)
    for i in range(0, size):
        in_combined[i] = in1[i] and in2[1]
    return np.array(in_combined)
```

```python
filter_tv = isin_extended(tv_index, [ radio_index, digital_index])
filter_radio = isin_extended(radio_index, [ tv_index, digital_index])
filter_digital = isin_extended(digital_index, [ tv_index, radio_index])
```

```python
tv_filtered = tv_index[filter_tv]
radio_filtered = radio_index[filter_radio]
digital_filtered = digital_index[filter_digital]
```

```python
print(tv_index)
print(tv_filtered)
print(radio_index)
print(radio_filtered)
print(digital_index)
print(digital_filtered)
```

## Models

### Model 1

```python
def marketing_boost(index, iterrations, discount, value):
    if discount**iterrations > 0.01 and index > 0:
        tv_marketing = tv_f * df['tv_grps'][index] * value
        radio_marketing = rd_f * df['radio_grps'][index] * value
        digital_marketing = dg_f * df['digital_grps'][index] * value
        marketing_addition = discount**iterrations * (tv_marketing + radio_marketing + digital_marketing)
        return marketing_addition + marketing_boost(index - 1, iterrations +1, discount, value)
    else:
        return 0
```

```python
def model_1():
    size = df['sales'].size
    predicted_values = []
    value = df['sales'][0]
    for i in range(size):
        pred_value = value + marketing_boost(i, 0, 0.9, value)
        predicted_values.append(pred_value)
    return predicted_values
            
```

```python
sales_pred = model_1()
```

```python
df_no_week['sales_pred'] = sales_pred
```

```python
def marketing_boost(index, iterrations, discount, value):
    if discount**iterrations > 0.01 and index > 0:
        tv_marketing = tv_f * df['tv_grps'][index] * value
        radio_marketing = rd_f * df['radio_grps'][index] * value
        digital_marketing = dg_f * df['digital_grps'][index] * value
        marketing_addition = discount**iterrations * (tv_marketing + radio_marketing + digital_marketing)
        return marketing_addition + marketing_boost(index - 1, iterrations +1, discount, value)
    else:
        return 0
```

```python
def temperature_effect(i, coef):
    value = coef * df['temp'][i]
    return value
```

```python
def trend(v, i):
    trend_multiplier = (df['sales'][259]/df['sales'][0]) ** (1/float(260))
    value = v * (trend_multiplier**i - 1)
    return value
```

```python
def model_2():
    size = df['sales'].size
    predicted_values = []
    value = df['sales'][0]
    for i in range(size):
        pred_value = value + marketing_boost(i, 0, 0.85, value) + temperature_effect(i, 3) + trend(value, i)
        predicted_values.append(pred_value)
    return predicted_values
            
```

### Model 3

#### To do list for next models

1. Create an error function for finetuning parameters
2. Improve the way how values are calculated. At the moment the new values are always calculated from the original value. (Moved to the next phase)
3. Finetune parameters with scipy library's minimize function

```python
tv_coef = tv_f
rd_coef= rd_f
dg_coef = dg_f
temp_coef = 3
trend_coef = (df['sales'][259]/df['sales'][0]) ** (1/float(260))
discount = 0.85
params = [tv_coef, rd_coef, dg_coef, temp_coef, trend_coef, discount]
```

```python
def marketing_boost(index, iterrations, discount, value, params):
    if discount**iterrations > 0.01 and index > 0:
        tv_marketing = params[0] * df['tv_grps'][index] * value
        radio_marketing = params[1] * df['radio_grps'][index] * value
        digital_marketing = params[2] * df['digital_grps'][index] * value
        marketing_addition = discount**iterrations * (tv_marketing + radio_marketing + digital_marketing)
        return marketing_addition + marketing_boost(index - 1, iterrations +1, discount, value, params)
    else:
        return 0
```

```python
def temperature_effect(i, coef):
    value = coef * df['temp'][i]
    return value
```

```python
def trend(v, i, coef):
    value = v * (coef**i - 1)
    return value
```

```python
def model_3(params):
    size = df['sales'].size
    predicted_values = []
    value = df['sales'][0]
    for i in range(size):
        pred_value = value + marketing_boost(i, 0, params[5], value, params[:3]) + temperature_effect(i, params[3]) + trend(value, i, params[4])
        predicted_values.append(pred_value)
    return predicted_values
            
```

```python
def mean_square_error(params):
    square_error = (df['sales'] - model_3(params)) ** 2
    return np.mean(square_error)
```

```python
Nfeval = 1
def callbackF(Xi):
    print(1)
    global Nfeval
    print '{0:4d}   {1: 3.6f}   {2: 3.6f}   {3: 3.6f}   {4: 3.6f}   {5: 3.6f}   {6: 3.6f}   {7: 3.6f}'.format(Nfeval, Xi[0], Xi[1], Xi[2], Xi[3], Xi[4], Xi[5], mean_square_error(Xi))
    Nfeval += 1
```

```python
print  '{0:4s}   {1:9s}   {2:9s}   {3:9s}   {4:9s}   {5:9s}   {6:9s}   {7:9s}'.format('Iter', "tv_coef", "rd_coef", "dg_coef", "temp_coef", "trend_coef", "discount", 'f(X)')   
result = optimize.minimize(mean_square_error, params, callback=callbackF,options={'disp': True} )
if result.success:
    fitted_params = result.x
    print(fitted_params)
else:
    raise ValueError(result.message)
```

Updating parameters according to findings of the minimize optimization

```python
sales_pred = model_3(params)
df_no_week['sales_pred'] = sales_pred
df_no_week.plot(figsize=(20,10), linewidth=5, fontsize=20)
```

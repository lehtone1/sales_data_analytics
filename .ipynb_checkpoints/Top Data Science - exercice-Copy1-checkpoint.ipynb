# TIme series analysis on sales data

## Get data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = 'simulated_sales.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
df.head()
```

## Modification and visualization

```python
df_no_week = df.loc[:, df.columns != 'week']
```

```python
df_no_week.update(df_no_week.iloc[:, 2:] /10)
```

```python
df_no_week
```

```python
df_no_week.plot(figsize=(20,10), linewidth=5, fontsize=20)
```

## Analysing data

### Correlation

```python
df_no_week.corr()
```

### Remove trend

```python
df_no_week_diff = df_no_week.diff()
```

```python
df_no_week_diff.plot(figsize=(20,10), linewidth=5, fontsize=20)
```

```python
df_no_week_diff.corr()
```

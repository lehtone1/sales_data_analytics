### Filtering double campaigns

```python
rule_tv = df['tv_grps'] != 0
rule_radio = df['radio_grps'] != 0
rule_digital = df['digital_grps'] != 0
```

```python
tv_index = df[rule_tv].index.values
radio_index = df[rule_radio].index.values
digital_index = df[rule_digital].index.values
```

```python
def isin_extended(listx1, listx2):
    in1 = np.isin(listx1, listx2[0], invert=True)
    in2 = np.isin(listx1, listx2[1], invert=True)
    size = listx1.size
    in_combined = np.ones(size, dtype=bool)
    for i in range(0, size):
        in_combined[i] = in1[i] and in2[1]
    return np.array(in_combined)
```

```python
filter_tv = isin_extended(tv_index, [ radio_index, digital_index])
filter_radio = isin_extended(radio_index, [ tv_index, digital_index])
filter_digital = isin_extended(digital_index, [ tv_index, radio_index])
```

```python
tv_filtered = tv_index[filter_tv]
radio_filtered = radio_index[filter_radio]
digital_filtered = digital_index[filter_digital]
```

```python
print(tv_index)
print(tv_filtered)
print(radio_index)
print(radio_filtered)
print(digital_index)
print(digital_filtered)
```
